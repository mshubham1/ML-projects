# Big Mart Sales

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train=pd.read_csv("train.csv")
test=pd.read_csv("test.csv")
```

```python
train.head()
```

```python
test.head()
```

```python
train['source']='train'
test['source']='test'
data = pd.concat([train, test],ignore_index=True)
print(train.shape, test.shape, data.shape)
```

```python
data.tail()
```

# Data Preprocessing

```python
data.info()
```

```python
data.describe()
```

```python
#missing values
data.apply(lambda x: sum(x.isnull()))
```

```python
# categiorical and numerical feature
categorical_columns = [x for x in data.dtypes.index if data.dtypes[x]=='object']
categorical_columns = [x for x in categorical_columns if x not in ['Item_Identifier','Outlet_Identifier','source']]
#Print frequency of categories
for col in categorical_columns:
    print('\nFrequency of Categories for varible %s'%col)
    print(data[col].value_counts())
```

```python
#imputing missing values
data['Item_Weight'].fillna(data['Item_Weight'].mean(), inplace = True)

data['Outlet_Size'].fillna(data['Outlet_Size'].mode()[0], inplace = True)

data['Item_Outlet_Sales'].fillna(data['Item_Outlet_Sales'].mean(), inplace = True)

data.isnull().sum()
```

```python
data.info()
```

```python
# combining reg, Regular and Low Fat, low fat and, L
data['Item_Fat_Content'] = data['Item_Fat_Content'].replace({'LF': 'Low Fat', 'reg': 'Regular', 'low fat': 'Low Fat'})
data['Item_Fat_Content'].value_counts()
```

```python
# Getting the first two characters of ID to separate them into different categories
data['Item_Identifier1'] = data['Item_Identifier'].apply(lambda x: x[0:2])
data['Item_Identifier1'] = data['Item_Identifier1'].map({'FD':'Food', 'NC':'Non_Consumable', 'DR':'Drinks'})
data['Item_Identifier1'].value_counts()
```

# Numerical and One hot Encoding of Categorical Feature

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['Outlet_Identifier1'] = le.fit_transform(data['Outlet_Identifier'])
var = ['Item_Fat_Content','Outlet_Location_Type','Outlet_Size','Item_Identifier1','Outlet_Type','Outlet_Identifier1']
le = LabelEncoder()
for i in var:
    data[i] = le.fit_transform(data[i])
```

```python
#One Hot Coding:
data = pd.get_dummies(data, columns=['Item_Fat_Content','Outlet_Location_Type','Outlet_Size','Outlet_Type','Item_Identifier1','Outlet_Identifier1'])
```

```python
data.head()
```

```python
data.dtypes
```

```python
import warnings
warnings.filterwarnings('ignore')
#Drop the columns which have been converted to different types:
data.drop(['Item_Type','Outlet_Establishment_Year'],axis=1,inplace=True)

#Divide into test and train:
train = data.loc[data['source']=="train"]
test = data.loc[data['source']=="test"]

#Drop unnecessary columns:
test.drop(['Item_Outlet_Sales','source'],axis=1,inplace=True)
train.drop(['source'],axis=1,inplace=True)

#Export files as modified versions:
train.to_csv("train_modified.csv",index=False)
test.to_csv("test_modified.csv",index=False)
```

# Model Building

```python
train2 = pd.read_csv("train_modified.csv")
test2 = pd.read_csv("test_modified.csv")
```

```python
train2.head()
```

```python
test2.head()
```

```python
X_train = train2.drop(['Item_Outlet_Sales', 'Outlet_Identifier','Item_Identifier'], axis=1)
y_train = train2.Item_Outlet_Sales
```

```python
X_test = test2.drop(['Outlet_Identifier','Item_Identifier'], axis=1)
```

```python
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
```

# Linear Regression

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
regressor.score(X_train, y_train)
```

```python
from sklearn import metrics
y_pred=regressor.predict(X_test)
```

```python
y_pred
```

```python
metrics.r2_score(y_train, regressor.predict(X_train))
```

```python
print('MAE:', metrics.mean_absolute_error(y_train, regressor.predict(X_train)))
print('MSE:', metrics.mean_squared_error(y_train, regressor.predict(X_train)))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_train, regressor.predict(X_train))))
```

```python
submission = pd.DataFrame({
'Item_Identifier':test2['Item_Identifier'],
'Outlet_Identifier':test2['Outlet_Identifier'],
'Item_Outlet_Sales': y_pred
},columns=['Item_Identifier','Outlet_Identifier','Item_Outlet_Sales'])
```

```python
submission.to_csv('submission_Linear_regression.csv',index=False)
```

```python
y_pred=randomforest_model.predict(X_test)
y_pred
```

```python
metrics.r2_score(y_train, randomforest_model.predict(X_train))
```

```python
print('MAE:', metrics.mean_absolute_error(y_train, randomforest_model.predict(X_train)))
print('MSE:', metrics.mean_squared_error(y_train, randomforest_model.predict(X_train)))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_train, randomforest_model.predict(X_train))))
```

```python
submission = pd.DataFrame({
'Item_Identifier':test2['Item_Identifier'],
'Outlet_Identifier':test2['Outlet_Identifier'],
'Item_Outlet_Sales': y_pred
},columns=['Item_Identifier','Outlet_Identifier','Item_Outlet_Sales'])
```
